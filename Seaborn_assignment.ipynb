# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# ANS:-

# Seaborn is a powerful Python visualization library built on top of Matplotlib. It offers a variety of plots for statistical data visualization. Here are five common plots and their uses:

#Scatter Plot:

#Uses: Scatter plots are used to visualize the relationship between two numerical variables. They help in identifying patterns, clusters, or trends in the data.

#Histogram:
#Uses: Histograms are useful for visualizing the distribution of a single numerical variable. They display the frequency of values within specified intervals, allowing for insights into the data's central tendency, dispersion, and skewness.

#Bar Plot:
##Uses: Bar plots are effective for comparing categorical variables and showing their distributions. They are commonly used to display counts, frequencies, or proportions of different categories.

#Box Plot (or Box-and-Whisker Plot):
## Uses: Box plots are excellent for visualizing the distribution and variability of numerical data. They display the median, quartiles, and potential outliers in the dataset, providing insights into the data's spread and central tendency.

#Heatmap:
##Uses: Heatmaps are useful for visualizing matrix-like data, especially correlation matrices. They use color gradients to represent the magnitude of values, making it easy to identify patterns, relationships, and clusters within the data

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

# ANS:-

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

# ANS:-

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1=sns.load_dataset("titanic")
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['age']=df1['age'].fillna(np.mean(df1['age']))
```

```python
df1.isnull().sum()
```

```python
sns.boxplot(x ='pclass',y = 'age' ,data=df1,color='g')
plt.show()
```

```python
sns.boxplot(x ='pclass',y = 'fare' ,data=df1,color='y')
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

# ANS:-

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df4=sns.load_dataset('flights')
```

```python
df4.info()
```

```python
df4.isnull().sum()
```

```python
df4['month'].unique()
```

```python
mapping={'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,"Jun":6,"Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}
```

```python
df4['month']=df4['month'].map(mapping)
```

```python
df4.head()
```

```python
df4.info()
```

```python
sns.heatmap(df4)
plt.show()
```
